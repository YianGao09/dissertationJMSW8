```python
import itertools
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd

# 假设 df2_with_dummies 已经是你的数据框

y = df2_with_dummies['hourseprice_change']
results = []

initial_vars = [col for col in df2_with_dummies.columns if 'borough_' in col]
additional_vars = ['disability_change', 'fertility_change', 'household_change', 'level4_change', 'populationdensity_change', 'employment_change']

# 逐步添加 additional_vars 并评估模型
for i in range(1, len(additional_vars) + 1):
    for combo in itertools.combinations(additional_vars, i):
        X = df2_with_dummies[initial_vars + list(combo)]
        X = sm.add_constant(X)
        X = X.astype(float)  # 确保所有数据都是浮点数类型
        model = sm.OLS(y, X).fit()
        results.append({
            'num_vars': i,
            'additional_vars': list(combo),
            'R2': model.rsquared,
            'Adj R2': model.rsquared_adj
        })

# 将结果转换为 DataFrame 并按 R2 降序排序
results_df = pd.DataFrame(results)
results_df_sorted = results_df.sort_values(by='R2', ascending=False).reset_index(drop=True)

# 输出结果
print("Additional Variables Combination and Corresponding R2 and Adjusted R2:\n")
print(results_df_sorted[['additional_vars', 'R2', 'Adj R2']])
```

```python
results_df_sorted[['additional_vars', 'R2', 'Adj R2']]
```

```python
# 计算平均 R² 和调整后的 R²
mean_results = results_df.groupby('num_vars').agg({
    'R2': 'mean',
    'Adj R2': 'mean'
}).reset_index()

# 打印平均结果
print(mean_results)
```

```python
# 绘制折线图
plt.figure(figsize=(10, 6))
plt.plot(mean_results['num_vars'], mean_results['R2'], marker='o', label='Mean R²')
plt.plot(mean_results['num_vars'], mean_results['Adj R2'], marker='o', label='Mean Adjusted R²')
plt.xlabel('Number of Additional Variables')
plt.ylabel('R² / Adjusted R²')
plt.title('Mean R² and Adjusted R² by Number of Additional Variables')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import statsmodels.api as sm

# 显示所有结果
print(results_df)

# 获取最佳模型的变量
best_model = results_df.loc[results_df['R2'].idxmax()]
print("Best model variables:", best_model['additional_vars'])
print("R2:", best_model['R2'])
print("Adj R2:", best_model['Adj R2'])

# 构建最终模型
X_best = df2_with_dummies[initial_vars + best_model['additional_vars']]
X_best = sm.add_constant(X_best)

# 确保所有的数据类型是float
X_best = X_best.astype(float)

# 进行模型拟合
model_best = sm.OLS(y, X_best).fit()

# 输出模型的详细总结
print(model_best.summary())
```

```python
import statsmodels.api as sm
import pandas as pd

# 假设results_df是包含所有结果的数据框
print(results_df)

# 获取最佳模型的变量
best_model = results_df.loc[results_df['R2'].idxmax()]
print("Best model variables:", best_model['additional_vars'])
print("R2:", best_model['R2'])
print("Adj R2:", best_model['Adj R2'])

# 构建最终模型
X_best = df2_with_dummies[initial_vars + best_model['additional_vars']]
X_best = sm.add_constant(X_best)

# 确保所有的数据类型是float
X_best = X_best.astype(float)

# 进行模型拟合
model_best = sm.OLS(y, X_best).fit()

# 获取p值并添加星号
p_values = model_best.pvalues
significance = []

for p in p_values:
    if p <= 0.001:
        significance.append('***')
    elif p <= 0.01:
        significance.append('**')
    elif p <= 0.05:
        significance.append('*')
    else:
        significance.append('')

# 将结果和星号合并
results_with_significance = pd.DataFrame({
    'coef': model_best.params,
    'std err': model_best.bse,
    't': model_best.tvalues,
    'P>|t|': model_best.pvalues,
    '[0.025': model_best.conf_int()[0],
    '0.975]': model_best.conf_int()[1],
    'significance': significance
})

print(results_with_significance)

# 输出模型的详细总结
print(model_best.summary())
```

```python

# 坐标信息Latitude	Longitud
u = cv2['lon']
v = cv2['lat']
g_coords = list(zip(u, v))
g_coords
```

```python
# z标准化：将数据转换为一个均值为0、标准差为1的正态分布。
g_X = (g_X - g_X.mean(axis=0)) / g_X.std(axis=0)
g_X

g_y = g_y.reshape((-1, 1))
g_y = (g_y - g_y.mean(axis=0)) / g_y.std(axis=0)
g_y
```

```python

## 4.2.3 拟合GWR模型
import libpysal as ps

#GWR相关库
from mgwr.gwr import GWR, MGWR
from mgwr.sel_bw import Sel_BW

import geopandas as gpd

#绘图库
import matplotlib.pyplot as plt
import matplotlib as mpl

#创建一个地理加权回归带宽选择器对象
gwr_selector = Sel_BW(g_coords, g_y, g_X)
gwr_selector
```

```python

#寻找最佳带宽
gwr_bw = gwr_selector.search(search_method='golden_section', criterion='AICc')

#search_method：指了搜索方法为黄金分割法
#指定用于评估模型好坏的标准，AICc（赤池信息准则）

print('最佳带宽大小为：', gwr_bw)
#进行GWR
gwr_results = GWR(g_coords, g_y, g_X, gwr_bw, fixed=False, kernel='bisquare', constant=True, spherical=True).fit()

gwr_results
```

```python
gwr_results.summary()
# 回归系数
gwr_results.params
# 整理回归系数
var_names = ['cof_Intercept', 'disability_change', 'fertility_change',
       'household_change', 'level4_change', 'populationdensity_change',
       'employment_change']
gwr_coefficent = pd.DataFrame(gwr_results.params, columns=var_names)
gwr_coefficent
```
